```python
import os
import torch
import bitsandbytes as bnb
from datasets import load_dataset
import transformers
from transformers import AutoTokenizer, AutoConfig, LlamaForCausalLM, LlamaTokenizer
from peft import prepare_model_for_kbit_training, LoraConfig, get_peft_model
from peft import PeftModel
```

```python
# Etapa 4: Definir parâmetros de treinamento
```

```python
MICRO_BATCH_SIZE = 4
BATCH_SIZE = 128
GRADIENT_ACCUMULATION_STEPS = BATCH_SIZE // MICRO_BATCH_SIZE
EPOCHS = 3
LEARNING_RATE = 3e-4
CUTOFF_LEN = 256
LORA_R = 8
LORA_ALPHA = 16
LORA_DROPOUT = 0.05
drive_path = '/content/drive/MyDrive/ADAN/ADAN_notebooks/LLM/input/'  # Ajuste o caminho conforme sua estrutura no Google Drive
```

```python
# Etapa 5: Carregar o modelo pré-treinado LLaMA e o tokenizador
```

```python
model = LlamaForCausalLM.from_pretrained('baffo32/decapoda-research-llama-7B-hf', load_in_8bit=True, device_map='auto')
tokenizer = LlamaTokenizer.from_pretrained('baffo32/decapoda-research-llama-7B-hf', add_eos_token=True)
model = prepare_model_for_kbit_training(model)
config = LoraConfig(r=LORA_R, lora_alpha=LORA_ALPHA, target_modules=['q_proj', 'v_proj'], lora_dropout=LORA_DROPOUT, bias='none', task_type='CAUSAL_LM')
model = get_peft_model(model, config)
tokenizer.pad_token_id = 0  # id do token de padding
```

```python
# Etapa 6: Preparar o dataset JSON
```

```python
def generate_prompt(data_point):
    return f"""Below is a document along with its respective author class. Use this information to learn how to associate documents with the correct author.
### Document:
Title: {data_point['title']}
Author: {data_point['author']}
Co-authors: {', '.join(data_point['coauthors'])}
Abstract: {data_point['abstract']}
### Author Class:
Author ID: {data_point['class']}"""

def tokenize(prompt):
    result = tokenizer(prompt, truncation=True, max_length=CUTOFF_LEN + 1, padding='max_length')
    return {'input_ids': result['input_ids'][:-1], 'attention_mask': result['attention_mask'][:-1]}

#json_folder_path =

data_file_path = os.path.join(drive_path, 'full_dataset_llm.json')
data = load_dataset('json', data_files=data_file_path)
data = data['train']
tokenized_data = data.map(lambda x: tokenize(generate_prompt(x)))
```

```python
# Etapa 7: Treinar o modelo
```

```python
trainer = transformers.Trainer(
    model=model,
    train_dataset=tokenized_data,
    args=transformers.TrainingArguments(
        per_device_train_batch_size=MICRO_BATCH_SIZE,
        gradient_accumulation_steps=GRADIENT_ACCUMULATION_STEPS,
        warmup_steps=100,
        num_train_epochs=EPOCHS,
        learning_rate=LEARNING_RATE,
        fp16=True,
        logging_steps=20,
        output_dir='/content/drive/MyDrive/ADAN/ADAN_notebooks/LLM/model',
        save_total_limit=3,
    ),
    data_collator=transformers.DataCollatorForLanguageModeling(tokenizer, mlm=False)
)
trainer.train(resume_from_checkpoint=False)


output_dir_model = '/content/drive/MyDrive/ADAN/ADAN_notebooks/LLM/model'

model_save_path = os.path.join(output_dir_model, 'lora-adan')
model.save_pretrained(model_save_path)
tokenizer.save_pretrained(model_save_path)
```
